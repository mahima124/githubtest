```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse as sp
from scipy.sparse.linalg import spsolve

def getAcc(pos, vel, q, m, Nx, boxsize, n0, Gmtx, Lmtx, B):
    N = pos.shape[0]
    dx = boxsize / Nx

    j = np.floor(pos / dx).astype(int)
    jp1 = np.mod(j + 1, Nx)
    weight_j = (jp1 * dx - pos) / dx
    weight_jp1 = (pos - j * dx) / dx
    n = np.bincount(j[:, 0], weights=weight_j[:, 0], minlength=Nx)
    n += np.bincount(jp1[:, 0], weights=weight_jp1[:, 0], minlength=Nx)
    n *= n0 * boxsize / N / dx

    phi_grid = spsolve(Lmtx, n - n0, permc_spec="MMD_AT_PLUS_A")
    E_grid = -Gmtx @ phi_grid
    E_grid =E_grid.reshape(N,1)
    #print(E_grid.shape)
    #print(vel.shape)
    F_Lorentz = q * (E_grid + np.cross(vel, B))

    a = F_Lorentz / m
    print(F_Lorentz)

    return a

def main():
    N = 4000 # Number of particles
    Nx = 4000   # Number of mesh cells
    t = 0
    tEnd = 500
    dt = 1
    boxsize = 50
    n0 = 1
    vb = 3.5
    vth = 1
    A = 0.1
    plotRealTime = True

    B = np.array([0, 0, 1])  # Magnetic field

    q = -1.0
    m = 1.0

    np.random.seed(42)
    pos = np.random.rand(N, 2) * boxsize
    vel = vth * np.random.randn(N, 3) + vb
    Nh = int(N / 2)
    vel[Nh:]*= 0
    vel[:,2] = 0
    vel[:,:2] += A * np.sin(2 * np.pi * pos[:,0] / boxsize)[:, np.newaxis]

    dx = boxsize / Nx
    e = np.ones(Nx)
    diags = np.array([-1, 1])
    vals = np.vstack((-e, e))
    Gmtx = sp.spdiags(vals, diags, Nx, Nx)
    Gmtx = sp.lil_matrix(Gmtx)
    Gmtx[0, Nx - 1] = -1
    Gmtx[Nx - 1, 0] = 1
    Gmtx /= (2 * dx)
    Gmtx = sp.csr_matrix(Gmtx)

    diags = np.array([-1, 0, 1])
    vals = np.vstack((e, -2 * e, e))
    Lmtx = sp.spdiags(vals, diags, Nx, Nx)
    Lmtx = sp.lil_matrix(Lmtx)
    Lmtx[0, Nx - 1] = 1
    Lmtx[Nx - 1, 0] = 1
    Lmtx /= dx**2
    Lmtx = sp.csr_matrix(Lmtx)

    acc = getAcc(pos, vel, q, m, Nx, boxsize, n0, Gmtx, Lmtx, B)

    Nt = int(np.ceil(tEnd / dt))

    fig = plt.figure(figsize=(6, 6), dpi=120)

    for i in range(Nt):
        vel += acc * dt / 2.0
        pos += vel[:,:2] * dt
        pos = np.mod(pos, boxsize)
        acc = getAcc(pos, vel, q, m, Nx, boxsize, n0, Gmtx, Lmtx, B)
        vel += acc * dt / 2.0
        t += dt

        if plotRealTime or (i == Nt - 1):
            plt.cla()
            plt.scatter(pos[:Nh,0], vel[:Nh,0], s=0.4, color='blue', alpha=0.5)
            plt.scatter(pos[Nh:,0], vel[Nh:,0], s=0.4, color='red', alpha=0.5)
            plt.axis([0, boxsize, -6, 6])
            plt.pause(0.001)

    plt.xticks(fontweight='bold', fontsize=20)
    plt.yticks(fontweight='bold', fontsize=20)
    plt.xlabel('x', fontweight='bold', fontsize=25)
    plt.ylabel('v', fontweight='bold', fontsize=25)
    plt.savefig('pic.png', dpi=240)
    plt.show()

    return 0

if __name__ == "__main__":
    main()
```
